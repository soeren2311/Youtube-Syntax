```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alter_x = [20, 23, 25, 28, 29, 35, 41, 47, 51, 58, 64, 67, 68, 70, 71, 73]
eink_L1 = [800, 920, 1260, 1210, 1980, 2890, 2650, 4000, 4100, 4520, 4480, 3100, 
          2320, 1730, 1400, 1380] 

eink_L2 = [1200, 1320, 1560, 1710, 1980, 3560, 3340, 3820, 4850, 4360, 4200, 
           3800, 2720, 2900, 2600, 2920]

eink_L3 = [750, 900, 1160, 1410, 1480, 2160, 3100, 2980, 3450, 3480, 3300, 
           2000, 1820, 1810, 1300, 1280]


x_index = np.arange(len(alter_x))
dist = 0.25

plot_1 = plt.figure(figsize=(15, 10))

plt.bar(x_index - dist, eink_L1, width=dist, color='r', label="Deutschland")
plt.bar(x_index, eink_L2, width=dist, color='k', label="Schweden")
plt.bar(x_index + dist, eink_L3, width=dist, color='y', label="Spanien")

plt.xlabel('Alter in Jahren')
plt.ylabel('Einkommen in â‚¬')

plt.xticks(ticks=x_index, labels=alter_x)
plt.title('Zusammenhang von Alter und Einkommen')
plt.grid(True)
plt.legend()
```
