## Matplotlib - Funktionenplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 8, 10)
x
```

```python
y1 = x
plt.figure(figsize=(8, 8))
plt.plot(x, y1)
```

```python
y2 = x**2
plt.figure(figsize=(8, 8))
plt.plot(x, y2)
```

```python
y3 = np.exp(x)

plt.figure(figsize=(10, 10))
plt.plot(x, y3)
plt.title("Plot der nat√ºrlichen Exponentialfunktion")
plt.xlabel("X-Werte")
plt.ylabel("Y-Werte")
```

```python
y4 = np.log2(x)
plt.figure(figsize=(10, 10))
plt.plot(x, y4)
plt.title("Plot der nat√ºrlichen Logarithmusfunktion")
plt.xlabel("X-Werte")
plt.ylabel("Y-Werte")
```
